**importing libraries**

```python
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from astropy.io import fits
```

**Accessing the data**

```python
fits_file = fits.open('f001a066.fits')
print(fits_file)
image_data = fits_file[0].data
print(image_data)
```

**Ploting the image**

```python
plt.figure()
plt.imshow(image_data, origin = 'lower')
plt.colorbar()
plt.show()
```

```python
plt.figure()
plt.imshow(image_data, origin = 'lower', cmap='Greys', norm=LogNorm())
plt.colorbar()
plt.show()
```

```python
centerofGalaxy = image_data[3000:4500, 3200:4200]
plt.figure()
plt.imshow(centerofGalaxy, origin = 'lower', cmap='Greys', norm=LogNorm())
plt.colorbar()
plt.show()
```
